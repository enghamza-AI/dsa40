```python
# CONCEPT: SLIDING WINDOW
# PROBLEM:  Given a list of integers nums and a number k, return the maximum sum of any k consecutive elements.
```

```python
def max_k_sum(nums, k):
    if k > len(nums) or k <= 0:
        return None

    window_sum = sum(nums[:k])
    max_sum = window_sum

    for i in range(k, len(nums)):
        window_sum = window_sum - nums[i - k] + nums[i]
        if window_sum > max_sum:
            max_sum = window_sum
    return max_sum
```

```python
print(max_k_sum([2,1,5,1,3,2], 3))
print(max_k_sum([1,4,2,9,7], 2))
```
