```python
# CONCEPT: ROTATING AN ARRAY
# PROBLEM: Rotate Right Given a list and a number k, return the list rotated right by k positions.
```

```python
def rotate_right(nums, k):
    if not nums:
        return nums
    k = k % len(nums)
    return nums[-k:] + nums[:-k]
```

```python
# Test Case 1: Normal case
print(rotate_right([1, 2, 3, 4, 5], 2))  # Output: [4, 5, 1, 2, 3]

# Test Case 2: k > len(nums)
print(rotate_right([1, 2, 3], 4))  # Output: [3, 1, 2] (4 % 3 = 1)

# Test Case 3: Single element
print(rotate_right([1], 5))  # Output: [1] (5 % 1 = 0)

# Test Case 4: Empty list
print(rotate_right([], 3))  # Output: []

# Test Case 5: k = 0
print(rotate_right([1, 2, 3], 0))  # Output: [1, 2, 3]

# Test Case 6: k = len(nums)
print(rotate_right([1, 2, 3], 3))  # Output: [1, 2, 3] (3 % 3 = 0)
```
