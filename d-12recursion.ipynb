```python
# CONCEPT: Recursion Refresher: Countdown & Buildup
# PROBLEM:  Write a recursive function to sum all numbers in a list without using loops or sum()
```

```python
def sum_list(nums, depth=0):
    # Print the current call with indentation to show depth
    print("  " * depth + f"sum_list({nums})", end=" -> ")

    # Base case: empty list
    if not nums:
        print("0 (Base Case)")
        return 0

    # Recursive case: first element + sum of the rest
    result = nums[0] + sum_list(nums[1:], depth + 1)
    print("  " * depth + f"{nums[0]} + {result - nums[0]} = {result}")
    return result

# Test Cases
print("--- Test Case 1: [1, 2, 3, 4] ---")
print("Final Result:", sum_list([1, 2, 3, 4]))

print("\n--- Test Case 2: [5] ---")
print("Final Result:", sum_list([5]))

print("\n--- Test Case 3: [] ---")
print("Final Result:", sum_list([]))
```
