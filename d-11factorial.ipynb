```python
# CONCEPT: RECURSION, FACTORIALS
# PROBLEM: Write a recursive function that returns n! (n factorial). 0! = 1, 1! = 1, 5! = 120.
```

```python
def factorial(n):
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers.")

    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
```

```python
print(factorial(0))
print(factorial(5))
print(factorial(6))
```
