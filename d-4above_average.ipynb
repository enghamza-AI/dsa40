```python
# CONCEPT: Two-Pass Thinking
'''
First Pass: Gather information (e.g., calculate the average).
Second Pass: Use that information to build the result (e.g., filter numbers greater than the average).

Time Complexity: Still O(n) (two separate loops over n items = 2n → simplified to O(n)).
'''

# problem: Return a new list containing only the numbers strictly greater than the average of the input list.

```

```python
def above_average(nums):
    if not nums:
        return []

    average = sum(nums) / len(nums)

    result = [num for num in nums if num > average]
    return result
```

```python
print(above_average([1,2,3,4,10]))
print(above_average([5,5,5]))
print(above_average([]))
```
