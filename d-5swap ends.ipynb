```python
# CONCEPT: Strings as Sequences
# Problem: Swap Ends
'''
 Given a string of at least 2 characters, 
 return the string with its first and last characters swapped. 
 If the string has 0 or 1 character, return it unchanged.

'''
```

```python
def swap_ends(s):
    if len(s) < 2:
        return s

    return s[-1] + s[1:-1] + s[0]
```

```python
print(swap_ends("HAMZA"))
print(swap_ends("HZ"))
print(swap_ends("H"))
```
