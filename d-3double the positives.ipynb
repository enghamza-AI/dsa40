```python
# day3: Concept = Mutation - modifying the original list (eg: nums[0] = nums[0] * 2)
# PROBLEM: Given a list of integers, return a new list where every positive number is doubled.
# Negative numbers and zeros remain unchanged.
```

```python
def double_positives(nums):
    result = []
    for num in nums:
        if num > 0:
            result.append(num * 2)

        else:
            result.append(num)

    return result
```

```python
print(double_positives([1, -2, 3, 0, 5]))
```

```python
print(double_positives([-1, -2]))
```

```python
# Day 003: Double the Positives

# Concept
# - Learned about **building new lists** vs **mutating the original**.
# - Used **list comprehensions** for concise code.

# Problem
# - Return a new list with positives doubled.

# Solution
# - Used `[num * 2 if num > 0 else num for num in nums]`.

## Time Complexity
# - O(n) time, O(n) space.
```
