```python
# CONCEPT: Big-O Complexity

# PROBLEM: Given a list of integers nums and a target, determine if any two distinct elements in nums sum to target. Return True or False.

```

```python
# Solution 1: Nested Loops O(n2)

def two_sum_slow(nums, target):
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):

            if nums[i] + nums[j] == target:
                return True
    return False
```

```python
print(two_sum_slow([2, 7, 11, 15], 9))
print(two_sum_slow([1 , 2, 3], 21))
```

```python
print(two_sum_slow([2, 11, 11, 7], 9))
```

```python
# Solution 2: Hash Set O(n)

def two_sum_fast(nums, target):
    seen = set()
    for num in nums:
        complement = target - num
        if complement in seen:
            return True
        seen.add(num)
    return False
```

```python
print(two_sum_fast([2, 11, 11, 7], 9))
```
