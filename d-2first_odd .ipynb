```python
# Goal: Return the index of the first odd number in a list. If no odd numbers exist, return -1.
# Concept: Linear Search
```

```python
# Approach
# Traverse the list (use a for loop).
# Check if each number is odd (use num % 2 != 0).
# Return the index of the first odd number.
# If no odd number is found, return -1.
```

```python
nums = [2,4,6,8,10]
def first_odd(nums):
    for index, num in enumerate(nums):

        if num % 2 != 0:
            return index

    return -1

print(first_odd(nums))
```
