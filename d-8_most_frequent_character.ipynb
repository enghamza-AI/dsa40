```python
# Concept: Frequency Counting
# Count how many times each character appears in a string.
# problem: Return the most frequent character in a string. If there’s a tie, return the character that appears first in the string.
```

```python
def most_frequent(s):
    frequency = {}

    for char in s:
        if char in frequency:
            frequency[char] += 1
        else:
            frequency[char] = 1

    max_count = -1
    result_char = s[0]

    for char in s:
        if frequency[char] > max_count:
            max_count = frequency[char]
            result_char = char

    return result_char
```

```python
print(most_frequent("HAMZA"))
print(most_frequent("DIHINI"))
print(most_frequent("HELLO"))
```
