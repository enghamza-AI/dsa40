```python
# CONCEPT: RECURSION: STRINGS
# PROBLEM: Reverse a String (Recursive)** Given a string, return its reverse using recursion. No slicing, no built-ins.
```

```python
def reverse(s, depth=0):
    # Print the current call with indentation to show depth
    print("  " * depth + f"reverse('{s}')", end=" -> ")

    # Base case: empty string
    if not s:
        print("'' (Base Case)")
        return ""

    # Recursive case: last character + reverse of the rest
    reversed_rest = reverse(s[1:], depth + 1)
    result = s[-1] + reversed_rest
    print("  " * depth + f"'{s[-1]}' + '{reversed_rest}' = '{result}'")
    return result

# Test Cases
print("--- Test Case 1: 'hello' ---")
print("Final Result:", reverse("hello"))

print("\n--- Test Case 2: 'a' ---")
print("Final Result:", reverse("a"))

print("\n--- Test Case 3: '' ---")
print("Final Result:", reverse(""))
```
