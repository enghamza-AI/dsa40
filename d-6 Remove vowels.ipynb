```python
# CONCEPT: string iteration & building
# PROBLEM: Return the input string with all vowels (a, e, i, o, u, A, E, I, O, U) removed.
```

```python
def remove_vowels(s):
    vowels = "aeiouAEIOU"
    result = ""
    for char in s:
        if char not in vowels:
            result += char
    return result
```

```python
print(remove_vowels("Hello World"))
print(remove_vowels("AEIOU"))
```

```python
# OPTION 2 : using list and join 
```

```python
def remove_vowelss(s):
    vowels = "aeiouAEIOU"
    result = []
    for char in s:
        if char not in vowels:
            result.append(char)
    return "".join(result)
```

```python
print(remove_vowelss("HamzaAEIOU"))
```
