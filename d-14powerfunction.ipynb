```python
# CONCEPT: Exponentiation by recursion.
# PPROBLEM:  Power** Implement `power(base, exp)` that returns base^exp. Use simple recursion (no fast-exponentiation yet — just get it working).
```

```python
def power(base, exp):
    if exp == 0:
        return 1
    return base * power(base, exp - 1)
```

```python
# Test Case 1: exp = 0
print(power(2, 0))  # Output: 1

# Test Case 2: exp = 1
print(power(3, 1))  # Output: 3

# Test Case 3: Normal case
print(power(2, 10))  # Output: 1024

# Test Case 4: Negative base
print(power(-2, 3))  # Output: -8

# Test Case 5: exp = 3
print(power(3, 3))  # Output: 27
```
