```python
#Concept: Palindromes

#problem: IS IT A PALINDROME?
# Return True if the input string is a palindrome (ignoring case and spaces), otherwise return False.
```

```python
# Approach 1: Reverse and compare (SIMPLEST)
def is_palindrome(s):
    cleaned = ''.join(char.lower() for char in s if char.isalpha())
    return cleaned == cleaned[::-1]
    
```

```python
print(is_palindrome("hamza"))
print(is_palindrome("racecar"))
```

```python
# Approach 2: Two-pointer technique (More efficient for large strings)
def is_palindrome2(s):
    cleaned = "".join(char.lower() for char in s if char.isalpha())
    left, right = 0, len(cleaned) - 1
    while left < right:
        if cleaned[left] != cleaned[right]:
            return False
        left += 1
        right -= 1
    return True
```

```python
print(is_palindrome2("hamza"))
print(is_palindrome2("racecar"))
```
