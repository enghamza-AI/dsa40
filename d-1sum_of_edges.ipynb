```python
# problem: Sum Of Edges

# Give a list of integers, return the sum of the first and last element
# .if the list has only one element, return it twice (since first and last digit are same)
# .if the list is empty, return 0
```

```python
nums = [3, 1, 4, 1, 5, 9]

def sum_of_edges(nums):
    if not nums:
        return 0
    if len(nums) == 1:
        return nums[0] * 2
    return nums[0] + nums[-1]

print(sum_of_edges(nums))

# Testing all other cases

print(sum_of_edges([3,1,4,1,5,9]))

print(sum_of_edges([7]))
print(sum_of_edges([]))
```
